# 数据清洗：去除后四列无用信息

```python
data_clean = data.dropna(axis=1)
data_clean.head(5)
```

# 数据清洗：对status_type和status_published进行编码

```python
labelEncoder = LabelEncoder()

#将状态类型更改为整数
labelEncoder.fit(data_clean['status_type'])
data_clean['status_type'] = labelEncoder.transform(data_clean['status_type'])

#对日期进行编码
labelEncoder.fit(data_clean['status_published'])
data_clean['status_published'] = labelEncoder.transform(data_clean['status_published'])

data_clean.head(5)
```

# 数据预处理：产生组合特征

```python
data_clean['new_reaction_count'] = data_clean.iloc[:,6:11].sum(axis=1)
data_clean['postive_reactions'] = data_clean.iloc[:,5:8].sum(axis=1)
data_clean['negative_reactions'] = data_clean.iloc[:,8:11].sum(axis=1)
```

```python
data_clean['interesting'] = (data_clean.num_comments+data_clean.num_shares)/data_clean.num_reactions
data_clean.replace([np.inf, -np.inf, np.nan], 0.0, inplace=True)
data_clean.head()
```

# 保存清洗完毕的数据

```python
data_clean.info()
```

```python
data_clean.to_csv("data_clean.csv", index=True)
```
