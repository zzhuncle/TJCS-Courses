# 库

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

# 两个圈

```python
X1, y1 = datasets.make_circles(n_samples=2000, factor=0.5, noise=0.05)
plt.scatter(X1[:, 0], X1[:, 1], marker='o')
plt.show()
```

# 两个月牙

```python
X2, y2 = datasets.make_moons(n_samples=1000, shuffle=True, noise=0.07, random_state=None)
plt.scatter(X2[:, 0], X2[:, 1], marker='o')
plt.show()
```

# 三个大点

```python
X3,y3 = datasets.make_blobs(n_samples=1000,n_features=2,centers=3,cluster_std=[2.0,3.0,2.0])
plt.scatter(X3[:,0],X3[:,1],marker='o')
plt.show()
```

# 三个小点

```python
X4,y4 = datasets.make_blobs(n_samples=900,n_features=2,centers=3,cluster_std=[1, 1, 1])
plt.scatter(X4[:,0],X4[:,1],marker='o')
plt.show()
```

# 三条线

```python
X5,y5 = datasets.make_classification(n_samples=1000,n_features=2,n_redundant=0,n_informative=2,n_clusters_per_class=1,n_classes=3)
plt.scatter(X5[:,0],X5[:,1],marker='o')
plt.show()
```

# 均匀分布

```python
X6 = np.random.rand(2000,2)
plt.scatter(X6[:,0],X6[:,1],marker='o')
plt.show()
```

# 分类

```python
from sklearn.cluster import KMeans
from sklearn.cluster import SpectralClustering
from sklearn.cluster import MiniBatchKMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import Birch
from sklearn.cluster import DBSCAN
from sklearn.cluster import MeanShift
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
```

# KMeans

```python
plt.figure(figsize=(5, 20))

plt.subplot(611)
km1 = KMeans(n_clusters=2, init='k-means++', max_iter=300, n_init=10, random_state=0)
result1 = km1.fit_predict(X1)
plt.scatter(X1[:,0], X1[:,1], marker='o', c=result1)
plt.title('KMeans')

plt.subplot(612)
km2 = KMeans(n_clusters=2, init='k-means++', max_iter=300, n_init=10, random_state=0)
result2 = km2.fit_predict(X2)
plt.scatter(X2[:,0], X2[:,1], marker='o', c=result2)

plt.subplot(613)
km3 = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
result3 = km3.fit_predict(X3)
plt.scatter(X3[:,0], X3[:,1], marker='o', c=result3)

plt.subplot(614)
km4 = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
result4 = km4.fit_predict(X4)
plt.scatter(X4[:,0], X4[:,1], marker='o', c=result4)

plt.subplot(615)
km5 = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
result5 = km5.fit_predict(X5)
plt.scatter(X5[:,0], X5[:,1], marker='o', c=result5)

plt.subplot(616)
km6 = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
result6 = km6.fit_predict(X6)
plt.scatter(X6[:,0], X6[:,1], marker='o', c=result6)

plt.show()
```

# SpectralClustering

```python
plt.figure(figsize=(5, 20))

plt.subplot(611)
sc1 = SpectralClustering(n_clusters = 2)
result1 = sc1.fit_predict(X1)
plt.scatter(X1[:,0], X1[:,1], marker='o', c=result1)
plt.title('SpectralClustering')

plt.subplot(612)
sc2 = SpectralClustering(n_clusters = 2)
result2 = sc2.fit_predict(X2)
plt.scatter(X2[:,0], X2[:,1], marker='o', c=result2)

plt.subplot(613)
sc3 = SpectralClustering(n_clusters = 3)
result3 = sc3.fit_predict(X3)
plt.scatter(X3[:,0], X3[:,1], marker='o', c=result3)

plt.subplot(614)
sc4 = SpectralClustering(n_clusters = 3)
result4 = sc4.fit_predict(X4)
plt.scatter(X4[:,0], X4[:,1], marker='o', c=result4)

plt.subplot(615)
sc5 = SpectralClustering(n_clusters = 3)
result5 = sc5.fit_predict(X5)
plt.scatter(X5[:,0], X5[:,1], marker='o', c=result5)

plt.subplot(616)
sc6 = SpectralClustering(n_clusters = 4)
result6 = sc6.fit_predict(X6)
plt.scatter(X6[:,0], X6[:,1], marker='o', c=result6)

plt.show()
```

# MiniBatchKMeans

```python
plt.figure(figsize=(5, 20))

plt.subplot(611)
mbkm1 = MiniBatchKMeans(init='k-means++', n_clusters=2)
result1 = mbkm1.fit_predict(X1)
plt.scatter(X1[:,0], X1[:,1], marker='o', c=result1)
plt.title('MiniBatchKMeans')

plt.subplot(612)
mbkm2 = MiniBatchKMeans(init='k-means++', n_clusters=2)
result2 = mbkm2.fit_predict(X2)
plt.scatter(X2[:,0], X2[:,1], marker='o', c=result2)

plt.subplot(613)
mbkm3 = MiniBatchKMeans(init='k-means++', n_clusters=3)
result3 = mbkm3.fit_predict(X3)
plt.scatter(X3[:,0], X3[:,1], marker='o', c=result3)

plt.subplot(614)
mbkm4 = MiniBatchKMeans(init='k-means++', n_clusters=3)
result4 = mbkm4.fit_predict(X4)
plt.scatter(X4[:,0], X4[:,1], marker='o', c=result4)

plt.subplot(615)
mbkm5 = MiniBatchKMeans(init='k-means++', n_clusters=3)
result5 = mbkm5.fit_predict(X5)
plt.scatter(X5[:,0], X5[:,1], marker='o', c=result5)

plt.subplot(616)
mbkm6 = MiniBatchKMeans(init='k-means++', n_clusters=4)
result6 = mbkm6.fit_predict(X6)
plt.scatter(X6[:,0], X6[:,1], marker='o', c=result6)

plt.show()
```
